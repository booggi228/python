### Problem:
How many Unique patterns that exist in the historical stock data set, based on
fluctuations in price.

```python
from numpy import vstack,array
from numpy.random import rand
from scipy.cluster.vq import kmeans,vq
from math import sqrt
from sklearn.cluster import KMeans
```

```python
# Let's find the optimal number of clusters and assign each stock to a specific cluster
# Calculate the average annual percentage return and volatilities over one year period
returns = data.pct_change().mean()*252
returns = pd.DataFrame(returns)
returns.columns = ['Returns']
returns.Volatility = data.pct_change().std()*sqrt(252)
```

```python
returns.Returns
```

```python
returns.Volatility
```

```python
# format the data into numpy array to feed into K-means algorithm
df = np.asarray([np.asarray(returns.Returns), np.asarray(returns.Volatility)]).T
```

```python
# Apply the KMeans clustering and plotting the curve

X = df
distortions =[]
for k in range(2,20):
    k_means = KMeans(n_clusters = k)
    k_means.fit(X)
    distortions.append(k_means.inertia_)

fig = plt.figure(figsize=(15,5))
plt.plot(range(2,20),distortions)
plt.grid(True)
plt.title('Elbow-Curve')
plt.show()
```

```python
from pylab import plot, show
# use 5 clusters
centroids,_= kmeans(df, 5)
# Assign each sample to a cluster
idx,_ = vq(df, centroids)

# plotting using numpy's logical indexing
plot(df[idx==0,0],df[idx==0,1],'ob',
     df[idx==1,0],df[idx==1,1],'oy',
     df[idx==2,0],df[idx==2,1],'or',
     df[idx==3,0],df[idx==3,1],'og',
     df[idx==4,0],df[idx==4,1],'om')
plot(centroids[:,0],centroids[:,1],'sg',markersize=8)
show()
```

### Problem:
Identify which all stocks are moving together and which all stocks are different from each other.

```python
new_data = data.copy()
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# convert data into numpy arrays
arr = new_data.values

# Scaling the values
scale = StandardScaler()
arr_scaled = scale.fit_transform(arr)
pca = PCA(n_components=150)
pca.fit(arr_scaled)

# the amount of variance that each PCA explains
var = pca.explained_variance_ratio_

# cumulative variance calculation
cum_var = np.cumsum(np.round(var, decimals=4)*100)
print(cum_var)
```

```python
# Plot the cumulative variance
plt.plot(cum_var);
```

```python
# let's consider 25 variables 
pca = PCA(n_components =25)
pca_tf = pca.fit_transform(arr_scaled)
print(pca_tf)
```

```python
print('Number of PCA :',len(pca.components_))
print(abs(pca.components_))
```

```python
components = pd.DataFrame(pca.components_, columns = new_data.columns)
components.head()
```
