```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN',
'londON_StockhOlm', 'Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
'12. Air France', '"Swiss Air"']})
```

```python
# 1
df.at[1, 'FlightNumber'] = 10055
df.at[3, 'FlightNumber'] = 10075
df['FlightNumber'] = df['FlightNumber'].astype(int)
df.head()
```

```python
# 2
df_temp = pd.DataFrame(df['From_To'].str.split('_',1).tolist(), columns = ['From','To'])
df_temp.head()
```

```python
# 3
df_temp['From'] = df_temp['From'].str.capitalize()
df_temp['To'] = df_temp['To'].str.capitalize()
df_temp.head()
```

```python
# 4
df.drop(['From_To'], axis=1, inplace=True)
data = pd.concat([df, df_temp], axis=1)
data.head()
```

```python
# 5
df2 = pd.DataFrame({"delay_1":[23, 47, 0, 0, 0], 
                    "delay_2":[0, 0, 0, 0, 0], 
                    "delay_3":[24, 43, 87, 0, 0], 
                    "delay_4":[13, 0, 0, 0, 0], 
                    "delay_5":[67, 32, 0, 0, 0]})
df2 = df2.replace(0, 'NaN')
df = pd.concat([data, df2], axis=1)
df = df.drop('RecentDelays', axis=1)
df.head()
```
